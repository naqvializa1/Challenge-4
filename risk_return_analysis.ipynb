```python
# Import the required libraries and dependencies
import pandas as pd
from pathlib import Path
import numpy as np
%matplotlib inline
```

##### Portfolio 1 - Step 1: Using the 60-day rolling window, the daily return data, and the S&P 500 returns, calculate the covariance. Review the last five rows of the covariance of the portfolio.

```python
# Calculate the covariance using a 60-day rolling window 
# Review the last five rows of the covariance data
# brk60rol = daily_returns.iloc[:,[3]].rolling(window=60)
# tgr60rol = daily_returns.iloc[:,[2]].rolling(window=60)
# sp60rol = daily_returns.iloc[:,[4]].rolling(window=60)

covariance_brk_spy = daily_returns["BERKSHIRE HATHAWAY INC"].rolling(window=60).cov(daily_returns['S&P 500'])
covariance_brk_spy.tail()
```
